* Matplotlib

```python
import matplotlib.pyplot as p
```

```python
l = [1,2,3,4,5]
l2 = [30,40,50,60,70]
p.plot(l,l2,"g-")
p.xlabel("l values")
p.ylabel('l2 values')
p.grid()
p.show()
```

```python
a = [i**2 for i in range(1,10)]
b = [i**3 for i in range(1,10)]
p.plot(a,b,"g+")
p.show()
```

```python
import numpy as np
distance = [30,40,50,60,70]
time = [30,40,50,60,70]
distance = np.array(distance)
time = np.array(time)
speed =distance/time
# speed = distance/time
p.plot(distance,speed,time)
p.show()
```

```python
elements = np.arange(1,100)
sqr = np.sqrt(elements)
p.plot(elements,sqr,"go")
p.show()
```

```python
lab = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
expl = (0, 0.1, 0.1, 0)
p.pie(sizes,labels = lab,autopct='%1.1f%%',explode =expl,shadow=True )
p.show()
```
